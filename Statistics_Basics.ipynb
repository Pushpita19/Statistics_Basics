 Q.2 What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.
 - The measure of central tendency help to summarize a datset with a single representative value.
 - The three most common are mean, median and mode.
 - Mean - It is simply the average of the data.
 - It is the sum of all values divided by the number of values.
 - Mean is used when the data distribution is symmetric, and no significant outlier is there.

```python
a = [1,2,3,4,5,6,7,8,9]
mean = sum(a)/len(a)
print(mean)
```

- Median is the middle number when the data is arranged in ascending or descending order.
- It is mainly used when the given data is skewed and there are outliers.

```python
import numpy as np
a = [5,7,2,9,8]
median = np.median(a)
print("Median",median)
```

- Mode is the highest occurance value in the dataset.
- It is used when the data is caterogical.
- It is also used to know the common occurance.

```python
import statistics

data = [10, 20, 20, 30, 40, 20, 30]
mode = statistics.mode(data)

print("Mode:", mode)
```

Q.6 Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
- Skewness is a measure of the symmetricty in the distribution of the data.
- It shows whether the data leans more towards the left or the right of the mean.
- For no skewness data, mean , median and mode lies on the same point.
- In positive skewness, most values are concentrated on the left and mean is greater than median and mode.
-For negative skewness data, most values are concentrated on the left and
mean < median < mode.
- In general, median is preferred over mean in skewed data.

```python
from scipy.stats import skew
d = [2,3,5,7,9,11]
skew(d)
```

Q.7  What is the interquartile range (IQR), and how is it used to detect outliers?
- The interquartile range is a measure of statistical dispersion .
- It shows how spread out the middle 50% of the data.
- It is commonly used in box plots and outlier detection.
- IQR = Q3 -Q1

where Q3 us 75th percentile and Q1 is 25th percentile.

- An observation is said to be outlier when it is out of the range of outlier thresholds.
- Threasholds are lowerfence = Q1 -1.5 * IQR and upperfence = Q3 +1.5 * IQR.

```python
import numpy as np
data = [10, 20, 20, 30, 40, 20, 30]
np.percentile(data,[0,25,50,75,100])

```

```python
#Q3 = 30 and Q1 = 20
IQR = 30-20
print(IQR)
```

Q.8  Discuss the conditions under which the binomial distribution is used.
- Binomial distribution is the 'n' bernoulli trial.
- It is a discrete probability distribution used to model the number of successes in a fixed number of independent trials where each trial has only two possible outcomes: Success and failure.
- Conditions under which the binomial distribution is used:
- Fixed no. of trials, only two possible outcomes.
- Outcome of each individual trial does not affect the outcome of the others.

Q.12  Provide an example dataset, calculate both covariance and correlation, and interpret the results.

```python
import numpy as np
import pandas as pd

#A data is given of 5 students - their study hour and marks they have obtained.
data = {
    'Hours_Studied': [2, 4, 6, 8, 10],
    'Marks_Obtained': [50, 55, 65, 70, 85]
}

df = pd.DataFrame(data)

#Covariance
covariance = df.cov().iloc[0, 1]

# Correlation
correlation = df.corr().iloc[0, 1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```
